```python
import numpy as np
import pandas as pd
```

```python
# cs=m*math+c
def gradesc(x,y):
    m,c=0,0
    n=len(x)
    alpha=1
    k=1
    i=0
    yp=0
    while True:
        yp=m*x+c
        k=(0.5/n)*sum(val**2 for val in (yp-y))
        print(k)
        md=(-1/n)*sum((yp-y)*m)
        cd=(-1/n)*sum(yp-y)
        m+=alpha*md
        c+=alpha*cd
        i+=1
    print(i)
```

```python
df=pd.read_csv("test_scores.csv")
ip=np.array(df.math)
op=np.array(df.cs)
gradesc(ip,op)
```
